```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.cm as cm
plt.gca(projection='3d')
x=np.linspace(-10,10,5)
y=np.linspace(-10,10,5)
z=np.linspace(-10,10,5)
xv,yv,zv=np.meshgrid(x,y,z)
plt.quiver(xv,yv,zv,-xv,-yv,-zv)
plt.savefig("vector3d.pdf")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.cm as cm
from pylab import rcParams
rcParams['figure.figsize']=5,3
#vector plot
x=np.linspace(-2,2,10)
y=np.linspace(-2,2,10)
xv,yv=np.meshgrid(x,y)
#meshgrid is a 2D grid:(xv,yv) provides the coordinates a mesh of Nx*Ny grid
plt.figure()
plt.quiver(xv,yv,-xv**3,-yv**3)
#quiver=array: At point (xv,yv), make vector(-xv,-yv)
```
